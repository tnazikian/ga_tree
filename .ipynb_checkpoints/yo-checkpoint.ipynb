```python
"""
This class constructs a candidate function represented as a binary expression tree
containing both unary and binary operators. Operations are recursively added up to 
a maximum depth specified by the user. Values of leaf nodes are random and can
either be a variable or a constant.

"""

class tree_generator():
    
    def __init__(self, variables, r=10):
        # Constrain legal operators
        self.unary_root_ops = ['sin', 'cos', 'tan', '-']
        self.bin_ops = ["+", "/", "*", "-", "**"]
        self.unary_ops = ["sin", "cos", "tan", "+", "/", "*", "-", "**"]
        self.unary_after_binary = ["sin", "cos", "tan"]
        self.unary_after_unary_ops = ["-", "sin", "cos", "tan"]
        self.variables = variables
        self.r = r

    def generate_tree(self, max_depth, parent, node_list, s):
        """returns a nested dict that represents a candidate function"""
        if s not in node_list:
            node_list.append(s) # add parent node to list of parent nodes
        
        if max_depth > 0:
            d = random.random() #randomly choose whether to add unary/binary branch
            if d > 0.5:
                #unary
                d = random.random()
                if d > 0.5:
                    if parent == "binary":
                        op = np.random.choice(self.unary_after_binary)
                    elif parent == "unary":
                        op = np.random.choice(self.unary_after_unary_ops)
                    out = unary(op, self.generate_tree(max_depth - 1, "unary", node_list, s + '_right.'))
                    
                else:
                    out = self.generate_tree(max_depth - 1, "unary", node_list, s)                
            else:
                #binary
                op = np.random.choice(self.bin_ops)
                out = binary(self.generate_tree(max_depth - 1, "binary", node_list, s + '_left.'), 
                              op, 
                              self.generate_tree(max_depth - 1, "binary", node_list, s + '_right.'))
            
#             node_list.append(s) # add parent node to list of parent nodes

        else:
            d = random.random()
            if d > 0.5:
                #use variable
                out = np.random.choice(self.variables)
            else:
                #use constant
                out = ("({})".format(
                    round(np.random.uniform(-1*self.r, self.r), 2)))     
        return out
        
    def init_tree(self, max_depth):
        """Initializes root node, then calls generate_tree"""
        d = random.random()
        node_list = []
        s1 = ""
        s2 = ""
        if d > 0.5:
            op = np.random.choice(self.unary_root_ops)
            out = unary(op, self.generate_tree(max_depth - 1, "unary", node_list, s1+'_right.'))
            return out, node_list
        else:
            op = np.random.choice(self.bin_ops)
            out = binary(self.generate_tree(max_depth - 1, "binary", node_list, s1+'_left.'), 
                          op, 
                          self.generate_tree(max_depth - 1, "binary", node_list, s2+'_right.'))
            return out, node_list
        
t = tree_generator(["(var)"])
s, nl1 = t.init_tree(3)
p.printTree(s)
k, nl2 = t.init_tree(3)
p.printTree(k)
# print(s)
# print(traverse(s))
# print(k)
# print(traverse(k))
#print(eval(traverse(s)))
print(s)
print(nl1)
print(k)
print(nl2)
```

```python
def _add(node, v):
    new = [v, [], []]
    if node:
        left, right = node[1:]
        if not left:
            left.extend(new)
        elif not right:
            right.extend(new)
        else:
            _add(left, v)
    else:
        node.extend(new)

def binary_tree(s):
    root = []
    for e in s:
        _add(root, e)
    return root

def traverse(n, order):
    if n:
        v = n[0]
        if order == 'pre':
            yield v
        for left in traverse(n[1], order):
            yield left
        if order == 'in':
            yield v
        for right in traverse(n[2], order):
            yield right
        if order == 'post':
            yield v
```

```python
from functools import reduce
def deep_get(dictionary, keys, default=None):
    """ By Yuda Prawira
    Source: https://stackoverflow.com/questions/25833613/python-safe-method-to-get-value-of-nested-dictionary"""
    return reduce(lambda d, key: d.get(key, default) if isinstance(d, dict) else default, keys.split(".")[:-1], dictionary)
```

```python
class PrintTree:
    """source: https://ebadahmadzadeh.wordpress.com/2014/02/26/how-to-print-a-dictionary-based-tree-in-python/"""
    def printTree(self, tree, d = 0):
        if (tree == None or len(tree) == 0):
            print("\t" * d, "-")
        else:
            for key, val in tree.items():
                if (isinstance(val, dict)):
                    print( "\t" * d, key)
                    self.printTree(val, d+1)
                else:
                    print ("\t" * d, key, str('(') + val + str(')'))
```
